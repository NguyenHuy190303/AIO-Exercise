### 5. Lấy các bộ phim từ 2010 tới 2015, với rating nhỏ hơn 6.0 nhưng lại có doanh thu thuộc top 5% trên toàn bộ dataset.

```python
# Bước 1: Xóa các hàng có giá trị NaN trong 'Revenue (Millions)' và sắp xếp
data_sorted = data.dropna(subset=['Revenue (Millions)']).sort_values(
    by='Revenue (Millions)', ascending=True)
print("Number of rows after dropna and sorting:", len(data_sorted))

# Bước 2: Lấy top 5% giá trị lớn nhất trong 'Revenue (Millions)'
top_5_percent_count = int(0.05 * len(data))
data_in_5_per = data_sorted.tail(top_5_percent_count)
print("Number of rows in top 5%:", len(data_in_5_per))

# Bước 3: Lọc dữ liệu với điều kiện về 'Year' và 'Rating'
filter1 = data_in_5_per.loc[(data_in_5_per['Year'] >= 2010) &
                            (data_in_5_per['Year'] <= 2015) &
                            (data_in_5_per['Rating'] < 6.0)]
print("Number of rows after applying all filters:", len(filter1))

# Hiển thị kết quả cuối cùng
filter1
```

### 10. Dealing with missing values - Filling:

```python
revenue_mean = data['Revenue (Millions)'].mean()

data.fillna({'Revenue (Millions)': revenue_mean}, inplace=True)
data.isnull().sum()
```

### apply() functions:

```python
def rating_category(x):
    r_category = ['Good', 'Average', 'Bad']
    if x >= 7.5:
        return r_category[0]
    elif x >= 6.0:
        return r_category[1]
    else:
        return r_category[2]


data['Rating_category'] = data['Rating'].apply(rating_category)

data
```
